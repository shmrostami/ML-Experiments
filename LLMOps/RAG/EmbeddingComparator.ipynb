```python
import os
from langchain_ollama import OllamaEmbeddings

os.environ.setdefault("no_proxy", "127.0.0.1,localhost")
os.environ.setdefault("HTTPX_NO_PROXY", "127.0.0.1,localhost")


# Create embeddings with Ollama
embeddings = OllamaEmbeddings(model="llama3.2:3b")
response = embeddings.embed_query("The sky is blue because of Rayleigh scattering")
print("Embedding length of llama:", len(response))

# Create embeddings with nomic-embed-text
embeddings = OllamaEmbeddings(model="nomic-embed-text")
response = embeddings.embed_query("The sky is blue because of Rayleigh scattering")
print("Embedding length of nomic:", len(response))
```
